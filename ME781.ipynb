```python

X=df.drop(['Check'], axis=1)
y=df.Check
```

```python
 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=100)
print(len(X_train),len(y_train),len(X_test),len(y_test))
```

```python

from sklearn.linear_model import LogisticRegression
model_1=LogisticRegression()
model_1.fit(X_train,y_train)
predictions_1=model_1.predict(X_test)
print(accuracy_score(y_test, predictions_1))
```

```python

from sklearn.ensemble import RandomForestClassifier
model_2 = RandomForestClassifier(n_estimators=50)
model_2.fit(X_train, y_train)
predictions_2=model_2.predict(X_test)
print(accuracy_score(y_test, predictions_2))
```

```python

from sklearn.tree import DecisionTreeClassifier
model_3 = DecisionTreeClassifier()
model_3.fit(X_train, y_train)
predictions_3 = model_3.predict(X_test)
print(accuracy_score(y_test, predictions_3))
```

```python

from sklearn.neighbors import KNeighborsClassifier
model_4 =KNeighborsClassifier()  
model_4.fit(X_train,y_train)
predictions_4=model_4.predict(X_test)
print(accuracy_score(y_test, predictions_4))
```

```python

from sklearn.svm import SVC
model_5 = SVC()  
model_5.fit(X_train,y_train)
predictions_5=model_5.predict(X_test)
print(accuracy_score(y_test, predictions_5))
```

```python

from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, predictions_1)
print("Confusion Matrix:")
print(confusion)
confusion = confusion_matrix(y_test, predictions_2)
print("Confusion Matrix:")
print(confusion)
confusion = confusion_matrix(y_test, predictions_3)
print("Confusion Matrix:")
print(confusion)
confusion = confusion_matrix(y_test, predictions_4)
print("Confusion Matrix:")
print(confusion)
confusion = confusion_matrix(y_test, predictions_5)
print("Confusion Matrix:")
print(confusion)
```

```python
from sklearn.model_selection import cross_validate
def classification_models(X, y):
    np.random.seed(42)
    
    scoring = ['accuracy', 'precision', 'recall', 'f1'] 
    model_names = []
    accuracy = []
    precision = []
    recall = []
    f1 = []
    
    classifiers = [
        ('LogisticRegression', LogisticRegression()),
        ('DecisionTreeClassifier', DecisionTreeClassifier()),
        ('RandomForestClassifier', RandomForestClassifier()),
        ('SVC', SVC()),
        ('KNeighborsClassifier', KNeighborsClassifier()),
    ]
    
    for model_name, model in classifiers:
        model_names.append(model_name)
        cv_results = cross_validate(model, X, y, scoring=scoring, cv=5)
        
        accuracy.append(cv_results['test_accuracy'].mean())
        precision.append(cv_results['test_precision'].mean())
        recall.append(cv_results['test_recall'].mean())
        f1.append(cv_results['test_f1'].mean())
    
    data = pd.DataFrame({
        'Model name': model_names, 
        'Accuracy': accuracy, 
        'Precision': precision, 
        'Recall': recall,
        'F1 Score': f1
    })
    
    return data

scores = classification_models(X, y)
print(scores)
```
